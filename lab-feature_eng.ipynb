```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
#Check the shape of the data
print(spaceship.shape)
```

**Check for data types**

```python
#Check the data types
print(spaceship.dtypes)
```

**Check for missing values**

```python
# Check for missing values
print("Missing values before removal:")
print(spaceship.isnull().sum())

# Remove rows with missing values
df = spaceship.dropna()

# Check for missing values after removal
print("\nMissing values after row removal:")
print(df.isnull().sum())

# Print the cleaned DataFrame
print("\nCleaned DataFrame (rows removed):")
df
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
# Transform Cabin to extract the first letter of the cabin {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}
df['Cabin'] = df['Cabin'].str[0]  # Extract the first letter
```

- Drop PassengerId and Name

```python
# Delate columns PassengerId y Name
df = df.drop(['PassengerId', 'Name'], axis=1)
```

- For non-numerical columns, do dummies.

```python
# Create dummies
df = pd.get_dummies(df, columns=['HomePlanet', 'CryoSleep', 'Cabin', 'Destination'], drop_first=True)

# Visualize dataframe
df
```

**Perform Train Test Split**

```python
#Check the data types
print(df.dtypes)
```

```python
features = df.drop(columns = ["Transported"])
target = df["Transported"]
```

```python
#Train test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
X_train.head()
```

```python
y_train.head()
```

**Normalization**

```python
normalizer = MinMaxScaler()
```

```python
#Fit to our training data
normalizer.fit(X_train)
```

Transforming our training and testing data

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#Use KNN predective model as Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=35)
```

```python
knn.fit(X_train_norm, y_train)
```

- Evaluate your model's performance. Comment it

```python
#Evaluate
knn.score(X_test_norm, y_test)
```
